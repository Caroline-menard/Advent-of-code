```python
with open("day13.txt") as file:
    test = file.read().split("\n")
test

data =[]
folding =[]
for i in test: 
    if len(i)<2:
        continue
    elif "fold" in i :
        i= i.split(" ")[-1]
        i= i.split("=")
        folding.append((i[0],int(i[1])))
    else: 
        data.append((int(i.split(",")[0]),int(i.split(",")[1])))
```

```python
data = set(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[8,8])
for (i,j) in data:
    plt.scatter(x=i,y=j,color= "black")
```

```python
def sym_x(x,data):
    to_move =[(i,j) for (i,j) in data if i>x]
    new_data=[]
    for (i,j) in data:
        if (i,j) not in to_move:
            new_data.append((i,j))
        else:
            gap = abs(x-i)
            new_data.append((x-gap,j))
    return set(new_data)

def sym_y(y,data):
    to_move =[(i,j) for (i,j) in data if j>y]
    new_data=[]
    for (i,j) in data:
        if (i,j) not in to_move:
            new_data.append((i,j))
        else:
            gap = abs(y-j)
            new_data.append((i,y-gap))
    return set(new_data)
```

```python
def get_visu(data):
    plt.figure(figsize=[16,2])
    for (i,j) in data:
        plt.scatter(x=i,y=-j,color= "black",s=40)
    plt.show()
```

```python
%%time
#part1
d = data.copy()
#get_visu(d)
for (axe,n) in folding:
    if axe=="x":
        d=sym_x(n,d)
    else:
        d=sym_y(n,d)
    break
    #get_visu(d)
len(d)
```

```python
%%time
#part2
d = data.copy()

for (axe,n) in folding:
    if axe=="x":
        d=sym_x(n,d)
    else:
        d=sym_y(n,d)
   
get_visu(d)
```
